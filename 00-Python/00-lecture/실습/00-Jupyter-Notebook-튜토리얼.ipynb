## 커멘드 모드 (커서)

- 커서가 "초록색"은 **입력 모드**
- 커서가 "파란색"은 **커멘드 모드** 입니다.

커멘드 모드에서는 셀의 추가, 삭제 등의 명령을 실행할 수 있습니다.

Esc를 눌러 커서의 색깔을 파란색으로 변경할 수 있습니다.

## 셀의 종류 (코드/텍스트)

코드 셀: 코드 실행을 위한 셀

텍스트 셀: 주석을 달거나, 설명 문구등을 달기 위한 텍스트 전용 셀

### 주석과 텍스트 셀의 차이

코드셀 설명입니다.

```python
print('안녕하세요')
```

```python
# 이렇게 주석을 달 수 있습니다.
print('안녕하세요')
```

이것은 텍스트 셀입니다

```python
# 주석입니다.
print('안녕하세요')

# 주석입니다.
print('안녕하세요')
```

### 셀의 삽입(추가)

상단의 "Insert" - "Insert Cell Above" 혹은 "Insert Cell Below"를 눌러서 삽입합니다

2. 코드 셀 아래에 삽입 (커멘드 모드 + B)


### 셀의 삭제

우측의 **휴지통 아이콘을 클릭**하여 삭제합니다.

**[셀 삭제 관련 단축키]**

### 셀의 타입 변경

## 실행과 출력

* **셀**: 실행의 단위
* **실행**: 재생 버튼을 클릭하여, 실행 (해당 셀 실행)
* **실행 순서**: *인터프리터 방식* 으로 실행하며, 순차적으로 메모리에 기억합니다.


**[실행 관련 단축키]**

1. 해당 셀을 실행하고 커서는 해당 셀  (Ctrl + Enter)
2. 해당 셀을 실행하고 커서는 다음 셀  (Shift + Enter)
3. 해당 셀을 실행하고 커서는 다음에 셀 삽입  (Alt + Enter)
